```python
from elf import (
    get_puzzle_input,
    submit_puzzle_answer,
)
```

```python
text = get_puzzle_input(2025, 5)
```

## part 1

```python
ranges = text.split('\n\n')[0].splitlines()
ranges = sorted(tuple(map(int, x.split('-'))) for x in ranges)
```

```python
combined = []

current_l = ranges[0][0]
current_r = ranges[0][1]

for l, r in ranges[1:]:
    if l > current_r:
        combined.append((current_l, current_r))
        current_l = l
        current_r = r
    elif l <= current_r:
        current_l = min(current_l, l)
        current_r = max(current_r, r)

combined.append((current_l, current_r))
```

```python
total = 0

for num in map(int, text.split('\n\n')[1].splitlines()):
    for l, r in combined:
        if l <= num <= r:
            total += 1
            break
```

```python
result = submit_puzzle_answer(2025, 5, 1, total)
print(result.is_correct, result.message)
```

## part 2

```python
total = 0

for l, r in combined:
    total += r - l + 1
```

```python
result = submit_puzzle_answer(2025, 5, 2, total)
print(result.is_correct, result.message)
```
