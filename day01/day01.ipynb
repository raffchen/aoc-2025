```python
text = [
    "L68",
    "L30",
    "R48",
    "L5",
    "R60",
    "L55",
    "L1",
    "L99",
    "R14",
    "L82",
]
```

## part 1

```python
pos = 50
count = 0

for line in text:
    direction, amount = line[0], int(line[1:])
    if direction == "L":
        pos = (pos - amount) % 100
    else:
        pos = (pos + amount) % 100

    if pos == 0:
        count += 1
```

```python
count
```

```python
result = submit_puzzle_answer(2025, 1, 1, count)
print(result.is_correct, result.message)
```

## part 2

```python
pos = 50
count = 0

for line in text:
    direction, amount = line[0], int(line[1:])
    if direction == "L":
        flipped = (100 - pos) % 100
        rounds = (flipped + amount) // 100
        pos = (pos - amount) % 100
    else:
        rounds, pos = divmod(pos + amount, 100)

    count += abs(rounds) 
```

```python
count
```

```python
result = submit_puzzle_answer(2025, 1, 2, count)
print(result.is_correct, result.message)
```
