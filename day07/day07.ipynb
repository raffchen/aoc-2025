```python
from elf import (
    get_puzzle_input,
    submit_puzzle_answer,
)
```

```python
text = get_puzzle_input(2025, 7)
```

## part 1

```python
split = 0

beams = set([text.splitlines()[0].index('S')])
arr = [line for line in text.splitlines()[2:] if not all(i == '.' for i in line)]

for line in arr:
    temp = set()
    while beams:
        beam = beams.pop()
        if line[beam] == '^':
            if beam - 1 >= 0:
                temp.add(beam - 1)
            if beam + 1 < len(line):
                temp.add(beam + 1)
            split += 1
        else:
            temp.add(beam)
    beams = temp
```

```python
result = submit_puzzle_answer(2025, 7, 1, split)
print(result.is_correct, result.message)
```

```python
mapped = {i: dict(v) for i, v in mapped.items()}
```

```python
def count_paths(tree):
    max_level = len(tree)

    dp = defaultdict(lambda: defaultdict(int))

    root = next(iter(tree[0].keys()))
    dp[0][root] = 1

    for level in range(max_level + 1):
        for node, children in tree[level].items():
            count = dp[level][node]
            for child in children:
                dp[level + 1][child] += count

    return sum(dp[max_level + 1].values())
```

```python
count_paths(mapped)
```

```python
result = submit_puzzle_answer(2025, 7, 2, count_paths(mapped))
print(result.is_correct, result.message)
```
