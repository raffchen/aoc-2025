## part 1

```python
l = []

for r in text.split(','):
    start, end = r.split('-')

    # it will always be a number with an even number of digits
    if len(start) == len(end) and len(start) % 2 == 1:
        continue

    for i in range(int(start), int(end)+1):
        if len(str(i)) % 2 == 1:
            continue
            
        i_str = str(i)
        if i_str == i_str[:len(i_str)//2] * 2:
            l.append(i)
```

```python
sum(l)
```

```python
result = submit_puzzle_answer(2025, 2, 1, sum(l))
print(result.is_correct, result.message)
```

## part 2

```python
l = []

for r in text.split(','):
    start, end = r.split('-')

    for i in range(int(start), int(end)+1):          
        i_str = str(i)

        for j in range(2, len(i_str)+1):
            if i_str == i_str[:len(i_str)//j] * j:
                l.append(i)
                break
```

```python
sum(l)
```

```python
result = submit_puzzle_answer(2025, 2, 2, sum(l))
print(result.is_correct, result.message)
```
