## part 1

```python
candidates = [0] * len(text)

for bank, line in enumerate(text):
    l, r = 0, 0

    while r < len(line) - 1:
        r += 1 

        if int(line[l]+line[r]) > candidates[bank]:
            candidates[bank] = int(line[l]+line[r])

        if line[r] > line[l]:
            l = r
```

```python
sum(candidates)
```

```python
result = submit_puzzle_answer(2025, 3, 1, sum(candidates))
print(result.is_correct, result.message)
```

## part 2

```python
import numpy as np
```

```python
arr = np.array([[int(i) for i in line] for line in text])
```

```python
def foo(line, digits):
    """
    The algorithm behind this is to get the max element while leaving enough digits for the rest of the number.
    So at the beginning, with 12 digits left, we'll look for the max of everything but the last 11 elements. 
    My mind blanked on doing this in vanilla Python, so I did it using numpy and np.argmax to find the max and its index.
    Once we get the max index, we use recursion to repeat this on the rest of the elements to the right of the max. 
    Once we're left with just 1 digit left, we can just grab the max. 
    """
    if digits == 1:
        return [np.max(line)]

    max_idx = int(np.argmax(line[:-(digits - 1)]))
    
    left = [line[max_idx]]
    right = foo(line[max_idx+1:], digits - 1)
    
    return np.concatenate((left, right))
```

```python
candidates = np.array(list(map(lambda x: foo(x, 12), arr)))
```

```python
candidates
```

```python
ans = np.sum(np.apply_along_axis(lambda x: int(''.join(x.astype(str))), 1, candidates))
```

```python
result = submit_puzzle_answer(2025, 3, 2, int(ans))
print(result.is_correct, result.message)
```
