```python
from elf import (
    get_puzzle_input,
    submit_puzzle_answer,
)
import numpy as np
```

```python
text = get_puzzle_input(2025, 4).splitlines()
```

```python
sample_text = """..@@.@@@@.
@@@.@.@.@@
@@@@@.@.@@
@.@@@@..@.
@@.@@@@.@@
.@@@@@@@.@
.@.@.@.@@@
@.@@@.@@@@
.@@@@@@@@.
@.@.@@@.@.
""".splitlines()
```

## part 1

```python
arr = np.array([[1 if i == '@' else 0 for i in line] for line in text])
```

```python
arr, arr.shape
```

```python
rows, cols = arr.shape

count = 0

for row in range(rows):
    for col in range(cols):
        if arr[row, col] == 1:
            neighbors = 0
            
            if row > 0 and col > 0:
                neighbors += arr[row-1, col-1] == 1
    
            if row > 0:
                neighbors += arr[row-1, col] == 1
    
            if row > 0 and col < cols - 1:
                neighbors += arr[row-1, col+1] == 1
    
            if col > 0:
                neighbors += arr[row, col-1] == 1
    
            if col < cols - 1:
                neighbors += arr[row, col+1] == 1
    
            if row < rows - 1 and col > 0:
                neighbors += arr[row+1, col-1] == 1
    
            if row < rows - 1:
                neighbors += arr[row+1, col] == 1
    
            if row < rows - 1 and col < cols - 1:
                neighbors += arr[row+1, col+1] == 1

            if neighbors < 4:
                count += 1
```

```python
count
```

```python
result = submit_puzzle_answer(2025, 4, 1, count)
print(result.is_correct, result.message)
```

## part 2

```python
arr = np.array([[1 if i == '@' else 0 for i in line] for line in text])
rows, cols = arr.shape

removed = 0

while True:
    remove = []
    
    for row in range(rows):
        for col in range(cols):
            if arr[row, col] == 1:
                neighbors = 0
                
                if row > 0 and col > 0:
                    neighbors += arr[row-1, col-1] == 1
        
                if row > 0:
                    neighbors += arr[row-1, col] == 1
        
                if row > 0 and col < cols - 1:
                    neighbors += arr[row-1, col+1] == 1
        
                if col > 0:
                    neighbors += arr[row, col-1] == 1
        
                if col < cols - 1:
                    neighbors += arr[row, col+1] == 1
        
                if row < rows - 1 and col > 0:
                    neighbors += arr[row+1, col-1] == 1
        
                if row < rows - 1:
                    neighbors += arr[row+1, col] == 1
        
                if row < rows - 1 and col < cols - 1:
                    neighbors += arr[row+1, col+1] == 1
    
                if neighbors < 4:
                    remove.append((row, col))

    if len(remove) == 0:
        break

    for row, col in remove:
        arr[row, col] = 0
        removed += 1
```

```python
removed
```

```python
result = submit_puzzle_answer(2025, 4, 2, removed)
print(result.is_correct, result.message)
```
