```python
from elf import (
    get_puzzle_input,
    submit_puzzle_answer,
)
import numpy as np
```

```python
text = get_puzzle_input(2025, 10)
```

## part 1

```python
total = 0

for line in text.splitlines():
    A = []
    
    lights, *buttons, joltages = line.split()
    
    for button in buttons:
        a = [0] * (len(lights) - 2)
        for i in button.strip('()').split(','):
            a[int(i)] = 1
        A.append(a)
        
    A = np.array(A).T
    b = np.array([0 if i == '.' else 1 for i in lights.strip('[]')])
    n, m = A.shape

    num_presses = []
    for mask in range(1<<m):
        x = np.array([(mask>>j)&1 for j in range(m)], dtype=int)
        if np.array_equal((A.dot(x) % 2), b % 2):
            num_presses.append(x.sum())

    total += min(num_presses)
```

```python
result = submit_puzzle_answer(2025, 10, 1, total)
print(result.is_correct, result.message)
```

## part 2

```python
from scipy.optimize import milp, LinearConstraint
```

```python
total = 0

for line in text.splitlines():
    A = []
    
    lights, *buttons, joltages = line.split()
    
    for button in buttons:
        a = [0] * (len(lights) - 2)
        for i in button.strip('()').split(','):
            a[int(i)] = 1
        A.append(a)
        
    A = np.array(A).T
    b = np.array([int(i) for i in joltages.strip('{}').split(',')])
    n, m = A.shape

    c = np.ones(m, dtype=int)
    constraints = LinearConstraint(A, b, b)
    res = milp(c=c, integrality=np.ones(m), constraints=constraints, bounds=None)

    total += res.fun
```

```python
result = submit_puzzle_answer(2025, 10, 2, int(total))
print(result.is_correct, result.message)
```
