```python
total = 0

current = 0
op_i = 0

for line in arr.T:
    joined_line = ''.join(line).strip()
    
    if joined_line == '':
        total += current
        current = 0
        op_i += 1
        continue

    num = int(joined_line)

    if current == 0:
        current += num
    else:
        if ops[op_i] == '*':
            current *= num
        else:
            current += num

total += current
```
