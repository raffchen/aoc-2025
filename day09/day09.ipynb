```python
from elf import (
    get_puzzle_input,
    submit_puzzle_answer,
)
from itertools import combinations
```

```python
text = get_puzzle_input(2025, 9)
```

## part 1

```python
points = [tuple(map(int, i.split(','))) for i in text.splitlines()]
```

```python
max_area = 0

for p1, p2 in combinations(points, 2):
    new_area = (abs(p1[0] - p2[0]) + 1) * (abs(p1[1] - p2[1]) + 1)
    if new_area > max_area:
        max_area = new_area
```

```python
result = submit_puzzle_answer(2025, 9, 1, max_area)
print(result.is_correct, result.message)
```

## part 2

```python
from shapely import Polygon, box
```

```python
polygon = Polygon(points)
```

```python
max_area = 0

for p1, p2 in combinations(points, 2):
    new_area = (abs(p1[0] - p2[0]) + 1) * (abs(p1[1] - p2[1]) + 1)
    
    x_min, x_max = min(p1[0], p2[0]), max(p1[0], p2[0])
    y_min, y_max = min(p1[1], p2[1]), max(p1[1], p2[1])
    rect = box(x_min, y_min, x_max, y_max)
    
    if new_area > max_area and polygon.covers(rect):
        max_area = new_area
```

```python
result = submit_puzzle_answer(2025, 9, 2, max_area)
print(result.is_correct, result.message)
```
