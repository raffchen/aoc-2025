```python
text = get_puzzle_input(2025, 12)
```

## part 1

```python
*shapes_, regions = text.split('\n\n')
```

```python
shapes = {int(i): np.array([[0 if k == '.' else 1 for k in m] for m in j.split()]) for i, j in [line.split(':') for line in shapes_]}
areas = {i: len(j[j == 1]) for i, j in shapes.items()}
```

```python
count = 0

for line in regions.splitlines():
    size, counts = line.split(': ')
    w, l = list(map(int, size.split('x')))
    total_area = w * l
    
    shape_counts = {i: c for i, c in enumerate(map(int, counts.split()))}
    line_area = [areas[i] * c for i, c in shape_counts.items()]

    diff = total_area - sum(line_area)

    # after checking the differences, 
    # it was either a very small negative number [-5, -1],
    # meaning that it could never fit another present
    # or it was a very large number (e.g. 400+),
    # so all the presents would likely fit without any rearranging
    if diff > 0:
        count += 1
```

```python
result = submit_puzzle_answer(2025, 12, 1, count)
print(result.is_correct, result.message)
```
